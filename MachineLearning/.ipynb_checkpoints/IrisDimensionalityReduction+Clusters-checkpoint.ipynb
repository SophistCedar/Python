```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#查看数据集内部的数据关系
#sns.pairplot(iris, hue='species', height=1.5);
```

```python
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
#划分训练集和数据集
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,
                                                random_state=1)
```

# 1-classification

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class高斯朴素贝叶斯模型
model = GaussianNB()                       # 2. instantiate model不需要传超参数
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
```

```python
#计算准确率
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

# 2-dimensionality reduction

```python
from sklearn.decomposition import PCA  # 1. Choose the model class
model = PCA(n_components=2)            # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                      # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(X_iris)         # 4. Transform the data to two dimensions
X_2D.shape
```

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False);
```

# 3-clusters

```python
from sklearn.mixture import GaussianMixture as GMM  # 1. Choose the model class高斯混合模型,导入的时候要大写
model = GMM(n_components=3,covariance_type='full')  # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                                   # 3. Fit to data. Notice y is not specified!
y_gmm = model.predict(X_iris)                       # 4. Determine cluster labels
```

```python
iris['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data=iris, hue='species',
           col='cluster', fit_reg=False);
```

# 4-Loading and visualizing the digits data

```python
from sklearn import datasets
digits = datasets.load_digits()
print(digits.images.shape)
#图像数据是一个三维数组:1797个样本，每个样本由一个8×8像素网格组成。
print(digits.target.shape)
print(digits.data.shape)
#将像素数组展平，这样我们就有了表示每个数字的像素值的长度-64数组。
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),transform=ax.transAxes, color='green')
```

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

```python
#manifold learning algorithm 流形学习算法对数据降维，到二维
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:, 0], data_projected[:, 1], c=digits.target,
            edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 9.5);#确定颜色的值
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB#高斯朴素贝叶斯模型
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, y_model)
mat
```

```python
ax = sns.heatmap(mat,square=True,annot=True,cbar=False)
#ax.set_ylim([9,0]) #可以注销此语句，将下面的ylim显示全面
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```python
fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

test_images = Xtest.reshape(-1, 8, 8)

for i, ax in enumerate(axes.flat):
    ax.imshow(test_images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(y_model[i]),
            transform=ax.transAxes,
            color='green' if (ytest[i] == y_model[i]) else 'red')
```
