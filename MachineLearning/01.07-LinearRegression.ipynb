# Linear Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

# Simple Linear Regression

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
print("Model slope:    ", model.coef_[0])
print("Model slope:    ", model.coef_)
print("Model intercept:", model.intercept_)
```

The ``LinearRegression`` estimator is much more capable than this, however—in addition to simple straight-line fits, it can also handle multidimensional linear models of the form
$$
y = a_0 + a_1 x_1 + a_2 x_2 + \cdots
$$
where there are multiple $x$ values.
Geometrically, this is akin to fitting a plane to points in three dimensions, or fitting a hyper-plane to points in higher dimensions.

The multidimensional nature of such regressions makes them more difficult to visualize, but we can see one of these fits in action by building some example data, using NumPy's matrix multiplication operator:

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])
print(X.shape,y.shape)
plt.scatter(X[:,0],y)
plt.scatter(X[:,1],y)
plt.scatter(X[:,2],y)
```

```python
model.fit(X, y)
print(model.intercept_)
print(model.coef_)
#这里𝑦数据由三个随机𝑥值,和线性回归系数用于构造恢复数据。
```

# Basis Function Regression
将线性回归应用于变量之间的非线性关系的一个技巧是根据基函数转换数据。
我们实际上所做的是将一维的 𝑥 值投影到更高的维度中，这样线性拟合就可以拟合 𝑥 和 𝑦 之间更复杂的关系。

# Polynomial basis functions

```python
from sklearn.preprocessing import PolynomialFeatures
x = np.array([2, 3, 4])
poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(x[:, None])
```

```python
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(7),
                           LinearRegression())
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

poly_model.fit(x[:, np.newaxis], y)
yfit = poly_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

# Gaussian basis functions
当然，其他基函数也是可能的。 例如，一个有用的模式是拟合一个不是多项式基的和，而是高斯基的和的模型。

```python
from sklearn.base import BaseEstimator, TransformerMixin

class GaussianFeatures(BaseEstimator, TransformerMixin):
    """Uniformly spaced Gaussian features for one-dimensional input"""
    
    def __init__(self, N, width_factor=2.0):
        self.N = N
        self.width_factor = width_factor
    
    @staticmethod
    def _gauss_basis(x, y, width, axis=None):
        arg = (x - y) / width
        return np.exp(-0.5 * np.sum(arg ** 2, axis))
        
    def fit(self, X, y=None):
        # create N centers spread along the data range
        self.centers_ = np.linspace(X.min(), X.max(), self.N)
        self.width_ = self.width_factor * (self.centers_[1] - self.centers_[0])
        return self
        
    def transform(self, X):
        return self._gauss_basis(X[:, :, np.newaxis], self.centers_,
                                 self.width_, axis=1)
    
gauss_model = make_pipeline(GaussianFeatures(20),
                            LinearRegression())
gauss_model.fit(x[:, np.newaxis], y)
yfit = gauss_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlim(0, 10);
```

# Regularization

```python
model = make_pipeline(GaussianFeatures(30),
                      LinearRegression())
model.fit(x[:, np.newaxis], y)

plt.scatter(x, y)
plt.plot(xfit, model.predict(xfit[:, np.newaxis]))

plt.xlim(0, 10)
plt.ylim(-1.5, 1.5);
```

```python
def basis_plot(model, title=None):
    fig, ax = plt.subplots(2, sharex=True)
    model.fit(x[:, np.newaxis], y)
    ax[0].scatter(x, y)
    ax[0].plot(xfit, model.predict(xfit[:, np.newaxis]))
    ax[0].set(xlabel='x', ylabel='y', ylim=(-1.5, 1.5))
    
    if title:
        ax[0].set_title(title)

    ax[1].plot(model.steps[0][1].centers_,
               model.steps[1][1].coef_)
    ax[1].set(xlabel='basis location',
              ylabel='coefficient',
              xlim=(0, 10))
    
model = make_pipeline(GaussianFeatures(30), LinearRegression())
basis_plot(model)
```

当基函数重叠时，这是典型的过拟合行为:相邻基函数的系数爆炸并相互抵消。我们知道这种行为是有问题的，如果我们能够通过惩罚模型参数的大值来明确地限制模型中的这种峰值，那就太好了。这种惩罚被称为正规化，有几种形式。

# Ridge regression ($L_2$ Regularization)

Perhaps the most common form of regularization is known as *ridge regression* or $L_2$ *regularization*, sometimes also called *Tikhonov regularization*.
This proceeds by penalizing the sum of squares (2-norms) of the model coefficients; in this case, the penalty on the model fit would be 
$$
P = \alpha\sum_{n=1}^N \theta_n^2
$$
where $\alpha$ is a free parameter that controls the strength of the penalty.
This type of penalized model is built into Scikit-Learn with the ``Ridge`` estimator:

```python
from sklearn.linear_model import Ridge
model = make_pipeline(GaussianFeatures(30), Ridge(alpha=0.1))
basis_plot(model, title='Ridge Regression')
```

$\alpha$参数本质上是一个控制结果模型复杂度的旋钮。
在$\alpha \to 0$的极限下，恢复标准线性回归结果;在$\alpha \to \infty$的限制中，所有模型响应都将被抑制。
脊回归的一个特别的优点是它可以被非常有效地计算——几乎不需要比原来的线性回归模型更多的计算成本。

# Lasso regression ($L_1$ regularization)

Another very common type of regularization is known as lasso, and involves penalizing the sum of absolute values (1-norms) of regression coefficients:
$$
P = \alpha\sum_{n=1}^N |\theta_n|
$$
Though this is conceptually very similar to ridge regression, the results can differ surprisingly: for example, due to geometric reasons lasso regression tends to favor *sparse models* where possible: that is, it preferentially sets model coefficients to exactly zero.

We can see this behavior in duplicating the ridge regression figure, but using L1-normalized coefficients:

```python
from sklearn.linear_model import Lasso
model = make_pipeline(GaussianFeatures(30), Lasso(alpha=0.001))
basis_plot(model, title='Lasso Regression')
```

使用lasso回归惩罚，大多数系数正好为零，功能行为由可用基函数的一个小子集建模。 与脊规范化一样， 𝛼 参数会调优惩罚的强度，并且应该通过交叉验证(例如)来确定

# Example: Predicting Bicycle Traffic
举个例子，让我们看看能否根据天气、季节和其他因素来预测骑自行车通过西雅图弗里蒙特大桥的次数。
在本节中，我们将把自行车数据与另一个数据集连接起来，并尝试确定天气和季节因素——温度、降水和日照时间——在多大程度上影响通过这条走廊的自行车交通量。 

```python
import pandas as pd
counts = pd.read_csv('data/Fremont_Bridge_Bicycle_Counter.csv', index_col='Date', parse_dates=True)
weather = pd.read_csv('data/BicycleWeather.csv', index_col='DATE', parse_dates=True)
```

```python
counts.head()
```

```python
#计算每天经过的自行车总数量
daily = counts.resample('d').sum()#原数据统计了每个小时的数量，此处降采样统计每天24h经过的总数量
daily['Total'] = daily.sum(axis=1)#再将Fremont Bridge Total	Fremont Bridge East Sidewalk	Fremont Bridge West Sidewalk三列数据按行计算总和
daily = daily[['Total']] # remove other columns
daily.head()
```

```python
#上面每天的自行车数量一直在变化;添加添加表示星期几的二进制列:
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
print((pd.date_range('1/1/2020', periods=7, freq='d').dayofweek))
for i in range(7):
    daily[days[i]] = (daily.index.dayofweek == i).astype(float)
daily.head()
```

```python
#增加是否是假日的指标
from pandas.tseries.holiday import USFederalHolidayCalendar
cal = USFederalHolidayCalendar()
holidays = cal.holidays('2012', '2016')
daily = daily.join(pd.Series(1, index=holidays, name='holiday'))
daily['holiday'].fillna(0, inplace=True)
daily.head()
```

```python
#增加白天的时间指标;用标准天文计算来补充这些信息:
def hours_of_daylight(date, axis=23.44, latitude=47.61):
    """Compute the hours of daylight for the given date"""
    days = (date - pd.datetime(2000, 12, 21)).days
    m = (1. - np.tan(np.radians(latitude))
         * np.tan(np.radians(axis) * np.cos(days * 2 * np.pi / 365.25)))
    return 24. * np.degrees(np.arccos(1 - np.clip(m, 0, 2))) / 180.

daily['daylight_hrs'] = list(map(hours_of_daylight, daily.index))#map(func, *iterables)使用来自iterables的参数应用func函数
daily[['daylight_hrs']].plot()
plt.ylim(8, 17)
```

```python
weather.head()
```

```python
#增加平均温度和总降水量到数据中。除了降水的英寸数之外，还可以添加一个标志，表示一天是否干燥(降水为零):
# temperatures are in 1/10 deg C; convert to C
weather['TMIN'] /= 10
weather['TMAX'] /= 10
weather['Temp (C)'] = 0.5 * (weather['TMIN'] + weather['TMAX'])

# precip is in 1/10 mm; convert to inches
weather['PRCP'] /= 254
weather['dry day'] = (weather['PRCP'] == 0).astype(int)

daily = daily.join(weather[['PRCP', 'Temp (C)', 'dry day']])
daily.head()
```

```python
#添加一个从第一天开始增加的计数器，并度量经过了多少年。
daily['annual'] = (daily.index - daily.index[0]).days / 365.
```

```python
daily.head()
```

```python
# Drop any rows with null values
daily.dropna(axis=0, how='any', inplace=True)

column_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun', 'holiday',
                'daylight_hrs', 'PRCP', 'dry day', 'Temp (C)', 'annual']
X = daily[column_names]
y = daily['Total']

model = LinearRegression(fit_intercept=False)
model.fit(X, y)
daily['predicted'] = model.predict(X)
daily[['Total','predicted']].head()#显示实际的降水量和预测的降水量之间的关系
```

```python
#画出实际降水量减去预测降水量的图
daily['bias']=daily['Total']-daily['predicted']
daily['bias'].plot(alpha=0.5);
```

```python
daily[['Total', 'predicted']].plot(alpha=0.5);
```

很明显，我们错过了一些关键的特征，尤其是在夏季。要么是我们的功能不完整(即在美国，人们决定骑车上班的依据不仅仅是这些)或者有一些我们没有考虑到的非线性关系(例如，也许人们在高温和低温下骑车的时间更少)。尽管如此，我们的粗略估计已经足够给我们一些启示，我们可以看看线性模型的系数来估计每个特征对每日自行车数量的影响:

```python
params = pd.Series(model.coef_, index=X.columns)
params
```

如果不衡量这些数字的不确定性，就很难解释它们。我们可以计算这些不确定性迅速使用bootstrap重振幅的数据:

```python
from sklearn.utils import resample
# resample(*arrays, **options) Resample arrays or sparse matrices in a consistent way
# Returns resampled_arrays : sequence of indexable data-structures. Sequence of resampled copies of the collections. 
    
np.random.seed(1)
# 生成列表[model.fit(*resample(X, y)).coef_ for i in range(1000)]
# 获取模型训练后的coef_，遍历1000次，放到列表中
# 计算列表内每一列的标准差
err = np.std([model.fit(*resample(X, y)).coef_
              for i in range(1000)], axis=0)
[model.fit(*resample(X, y)).coef_ for i in range(1000)][:2]
```

```python
print(pd.DataFrame({'effect': params.round(0),
                    'error': err.round(0)}))
```

我们首先看到的是，在每周的基线中有一个相对稳定的趋势:工作日的乘客比周末和假期的乘客要多得多。我们看到，每增加一个小时的日照，就有129人选择骑车;气温每升高1摄氏度，就会有65人抢着骑自行车;干燥的一天意味着平均多54833名乘客，每一英寸的降水意味着多665 62人把自行车留在家里。一旦所有这些影响都被考虑在内，我们就会看到每年新增2718名每日骑行者。

我们的模型几乎肯定遗漏了一些相关信息。例如，非线性效应(如降水和冷温度的影响)和每个变量内的非线性趋势(如在非常冷和非常热的温度下不愿骑行)在这个模型中不能被解释。此外,我们已经丢弃的一些更细粒度的信息(比如一个下雨的早晨的区别和一个下雨的下午),我们忽略了天之间的相关性(例如一个下雨的可能影响周二周三的数据,或一个意想不到的影响经过连续的晴天雨天)。这些都是潜在的有趣的效果，如果您愿意，现在您已经有了开始探索它们的工具
