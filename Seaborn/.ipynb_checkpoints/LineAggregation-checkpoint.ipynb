```python
#模块导入
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
#更复杂的数据集对于相同的x变量值会有多个测量值。seaborn的默认行为是通过绘制均值和均值周围的95%置信区间来聚合每个x值的多个测量值:
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
#ci=None取消置信区间confidence interval
#置信区间是使用bootstrapping来计算的，对于较大的数据集来说，bootstrapping需要大量的时间。
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

```python
#通过绘制标准差而不是置信区间来表示分布在每个时间点的扩展:
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

```python
#为了完全关闭聚合，将estimator参数设置为None。当数据在每个点有多个观察值时，这可能会产生一种奇怪的效果。
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri);
```

```python
fmri.head()
```

```python
#在lineplot()中使用语义hue将确定如何聚合数据。
#添加一个具有两个级别的hue语义，将绘图分为两行和错误频带，每一行使用不同的颜色表示它们对应的数据子集。
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

```python
#使用style语义改变lineplot()内的破折号样式
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

```python
#突出标记每个点
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri);
```

```python
#尽量统一语义使用，这样看起来简单，比如将hue和style都指定为event变量
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri);
```

```python
print(fmri.shape)
print(fmri.query("event == 'stim'").shape)
print(fmri.query("event == 'stim'").head())
print(fmri.query("event == 'stim'")["subject"].unique())
```

```python
#单独绘制每个采样单元
sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=fmri.query("event == 'stim'"));
```

```python
dots = sns.load_dataset("dots")
dots.shape
```

```python
dots.head()
```

```python
dots=dots.query("align == 'dots'")
dots.shape
```

```python
dots.head()
```

```python
#hue可以分类也可以数字指定
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            kind="line", data=dots);
```

```python
#更改色调
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            palette=palette,
            kind="line", data=dots);
```

```python
#更改色调
from matplotlib.colors import LogNorm
palette = sns.cubehelix_palette(light=.7, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            hue_norm=LogNorm(),
            kind="line", data=dots);
```

```python
#size控制线宽
sns.relplot(x="time", y="firing_rate",
            size="coherence", style="choice",
            kind="line", data=dots);
```
