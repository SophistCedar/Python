```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.set(color_codes=True)
```

Plotting univariate distributions
The most convenient way to take a quick look at a univariate distribution in seaborn is the distplot() function. 
By default, this will draw a histogram and fit a kernel density estimate (KDE).

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
sns.distplot(x, kde=False)
```

```python
sns.distplot(x,kde=False,rug=True)
```

```python
sns.distplot(x, bins=20, kde=False, rug=True);
```

```python
sns.distplot(x, hist=False, rug=True);
```

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma);
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
sns.jointplot(x="x", y="y", data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

# Visualizing pairwise relationships in a dataset

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris);
```

```python
sns.pairplot(iris, hue="species");
#Specifying the hue parameter automatically changes the histograms to KDE plots to facilitate comparisons between multiple distributions.
```
