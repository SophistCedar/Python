```python
#模块导入
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
#查看Seaborn库中的所有可用数据集
sns.get_dataset_names()
```

```python
#加载数据
tips = sns.load_dataset("tips")
tips.shape
#加载数据时出现EmptyDataError: No columns to parse from file错误。
#是因为下载tips.csv文件时候中断了，导致文件内容为空，C:\Users\Administrator\seaborn-data在这个目录下可以查看下载的数据。
#解决办法：1-删除空文件后重新调用语句下载，并稍等一会再执行下面语句。也可到https://github.com/mwaskom/seaborn-data网站下载数据放到目录。
```

```python
tips.head()
```

```python
#画出普通二维散点图
#replot可以画直线或者散点图，通过指定kind实现
#kind="scatter"是默认的，可以指定kind="line"画直线
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
#hue="smoker",通过此smoker变量指定色彩，数据里的变量只有两个值Yes或者No，故只有两种颜色。
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
#使用style属性改变标记样式，体现不同类别之间的差异
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

```python
#使用time变量确定style
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

```python
#使用size变量确定hue色彩属性
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```python
#使用palette来自定义调色板
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

```python
#使用size改变每个点的大小
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

```python
#size对应的变量的值并不是点的面积大小，变量范围被规范化为面积单位范围，可以通过sizes自定义
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```
